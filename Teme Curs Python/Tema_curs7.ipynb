```python
# 2.
# ABSTRACTION
from abc import ABC, abstractmethod

class FormaGeometrica(ABC):
    def __init__(self):
        self.pi = 3.14
    @abstractmethod
    def aria(self):
        pass
    def descrie(self):
        print(f'Cel mai probabil am colturi')
        
# INHERITANCE

class Patrat(FormaGeometrica):
    def __init__(self, latura):
        super().__init__()
        self.__latura = latura #protected
        print(f'In clasa patrat pi = {self.pi}')
    def aria(self):
        return self.__latura**2
    
# ENCAPSULATION

    @property
    def latura(self):
        return self.__latura
    def set_latura(self, latura_noua):
        self.__latura = latura_noua
    @latura.getter
    def get_latura(self):
        return self.__latura
    @latura.deleter
    def del_latura(self):
        self.__latura = None
        print(f'Latura a fost stearsa')
        
class Cerc(FormaGeometrica):
    def __init__(self, raza):
        super().__init__()
        self.__raza = raza #protected
        print(f'Raza cercului este pi = {self.pi}')
    def aria(self):
        return self.pi*(self.__raza**2)
    @property
    def raza(self):
        return self.__raza
    def set_raza(self, raza_noua):
        self.__raza = raza_noua
    @raza.getter
    def get_raza(self):
        return self.__raza
    @raza.deleter
    def del_raza(self):
        self.__raza = None
        print(f'Raza a fost stearsa')
        
# POLYMORPHISM
    def descrie(self):
        print(f'Eu nu am colturi')

def descriere_noua(obj):
    obj.descrie()


p1 = Patrat(5)
p1.descrie()
descriere_noua(p1)
print(f'Latura este {p1.latura}')
print(f'Aria este {p1.aria()}')
p1.set_latura(7)
print(f'Noua latura este {p1.get_latura}')
del p1.del_latura

c1 = Cerc(4)
c1.descrie()
descriere_noua(c1)
print(f'Raza este {c1.raza}')
print(f'Aria este {c1.aria()}')
c1.set_raza(9)
print(f'Noua raza este {c1.get_raza}')
del c1.del_raza    

    
   
    
```
